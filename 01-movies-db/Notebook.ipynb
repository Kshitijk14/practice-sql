```python
import pandas as pd

df = pd.read_excel('movies-db.xlsx', sheet_name='financials')
df
```

```python
# Convert revenue to millions based on unit
def convert_to_millions(row):
    if row['unit'] == 'Billions':
        return row['revenue'] * 1000
    elif row['unit'] == 'Thousands':
        return row['revenue'] / 1000
    else:
        return row['revenue']

df['revenue_mils'] = df.apply(convert_to_millions, axis=1)
df
```

```python
# Convert revenue to INR
def convert_to_inr(row):
    if row['currency'] == 'USD':
        return row['revenue_mils'] * 83.95
    else:
        return row['revenue_mils']

df['revenue_inr_mils'] = df.apply(convert_to_inr, axis=1)
df
```

```python
# Calculate profit in INR in millions
df['profit_inr_mils'] = df['revenue_inr_mils'] - df['budget']
df
```
